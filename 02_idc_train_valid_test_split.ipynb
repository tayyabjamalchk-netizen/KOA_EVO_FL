```python
# ==========================================================
# SECTION 2 : DATASET PATH
# ==========================================================

DATASET_PATH = "/content/IDC_regular_ps50_idx5"

print("="*70)
print("DATASET PATH")
print("="*70)

print(DATASET_PATH)
print()

print("Exists :", os.path.exists(DATASET_PATH))
```

```python
# ==========================================================
# SECTION 3 : READ ALL IMAGES
# ==========================================================

image_paths = []
labels = []

patients = sorted(os.listdir(DATASET_PATH))

for patient in patients:

    patient_path = os.path.join(DATASET_PATH, patient)

    if not os.path.isdir(patient_path):
        continue

    for cls in ["0", "1"]:

        class_path = os.path.join(patient_path, cls)

        if not os.path.exists(class_path):
            continue

        for img in os.listdir(class_path):

            if img.endswith(".png"):

                image_paths.append(
                    os.path.join(class_path, img)
                )

                labels.append(int(cls))

print("="*70)
print("TOTAL IMAGES :", len(image_paths))
print("="*70)
```

```python
# ==========================================================
# SECTION 4 : CREATE DATAFRAME
# ==========================================================

df = pd.DataFrame({

    "filepath": image_paths,

    "label": labels

})

print("="*70)

print(df.head())

print("="*70)

print(df.shape)
```

```python
# ==========================================================
# SECTION 5 : SPLIT DATASET
# ==========================================================

train_df, temp_df = train_test_split(

    df,

    test_size=0.30,

    stratify=df["label"],

    random_state=42

)

valid_df, test_df = train_test_split(

    temp_df,

    test_size=0.50,

    stratify=temp_df["label"],

    random_state=42

)

print("="*70)

print("TRAIN :", len(train_df))

print("VALID :", len(valid_df))

print("TEST  :", len(test_df))
```

```python
# ==========================================================
# SECTION 6 : CREATE DATAFRAME
# ==========================================================

import pandas as pd

image_paths = []
labels = []

patients = sorted(os.listdir(DATASET_PATH))

for patient in patients:

    patient_path = os.path.join(DATASET_PATH, patient)

    if not os.path.isdir(patient_path):
        continue

    for cls in ["0", "1"]:

        class_path = os.path.join(patient_path, cls)

        if not os.path.exists(class_path):
            continue

        for img in os.listdir(class_path):

            if img.endswith(".png"):

                image_paths.append(
                    os.path.join(class_path, img)
                )

                labels.append(cls)

df = pd.DataFrame({
    "filename": image_paths,
    "class": labels
})

print("="*70)
print(df.head())
print("="*70)
print("Total Images :", len(df))
```

```python
# ==========================================================
# SECTION 7 : TRAIN VALID TEST SPLIT
# ==========================================================

from sklearn.model_selection import train_test_split

train_df, temp_df = train_test_split(
    df,
    test_size=0.30,
    stratify=df["class"],
    random_state=42
)

valid_df, test_df = train_test_split(
    temp_df,
    test_size=0.50,
    stratify=temp_df["class"],
    random_state=42
)

print("="*70)
print("TRAIN :", len(train_df))
print("VALID :", len(valid_df))
print("TEST  :", len(test_df))
```

```python
# ==========================================================
# SECTION 8 : IMAGE DATA GENERATOR
# ==========================================================

from tensorflow.keras.preprocessing.image import ImageDataGenerator

train_datagen = ImageDataGenerator(

    rescale=1./255,

    rotation_range=20,

    width_shift_range=0.2,

    height_shift_range=0.2,

    zoom_range=0.2,

    shear_range=0.2,

    horizontal_flip=True,

    fill_mode="nearest"

)

valid_datagen = ImageDataGenerator(
    rescale=1./255
)

test_datagen = ImageDataGenerator(
    rescale=1./255
)

print("="*70)
print("GENERATORS CREATED")
print("="*70)
```

```python
# ==========================================================
# SECTION 9 : TRAIN GENERATOR
# ==========================================================

train_generator = train_datagen.flow_from_dataframe(

    dataframe=train_df,

    x_col="filename",

    y_col="class",

    target_size=(224,224),

    batch_size=60,

    class_mode="binary",

    shuffle=True,

    seed=42

)

print("="*70)
print("TRAIN GENERATOR CREATED")
print("="*70)
```

```python
# ==========================================================
# SECTION 10 : VALIDATION GENERATOR
# ==========================================================

valid_generator = valid_datagen.flow_from_dataframe(

    dataframe=valid_df,

    x_col="filename",

    y_col="class",

    target_size=(224,224),

    batch_size=60,

    class_mode="binary",

    shuffle=False

)

print("="*70)
print("VALIDATION GENERATOR CREATED")
print("="*70)
```

```python
# ==========================================================
# SECTION 11 : TEST GENERATOR
# ==========================================================

test_generator = test_datagen.flow_from_dataframe(

    dataframe=test_df,

    x_col="filename",

    y_col="class",

    target_size=(224,224),

    batch_size=60,

    class_mode="binary",

    shuffle=False

)

print("="*70)
print("TEST GENERATOR CREATED")
print("="*70)
```

```python
# ==========================================================
# SECTION 12 : CLASS INDICES
# ==========================================================

print("="*70)
print("CLASS INDICES")
print("="*70)

print(train_generator.class_indices)
```

```python
# ==========================================================
# SECTION 13 : DATASET SUMMARY
# ==========================================================

print("="*70)
print("TRAIN IMAGES :", train_generator.samples)
print("VALID IMAGES :", valid_generator.samples)
print("TEST IMAGES  :", test_generator.samples)
print("BATCH SIZE   :", train_generator.batch_size)
print("="*70)
```

```python
# ==========================================================
# SECTION 14 : SAVE SPLITS
# ==========================================================

train_df.to_csv("train_split.csv", index=False)
valid_df.to_csv("valid_split.csv", index=False)
test_df.to_csv("test_split.csv", index=False)

print("CSV Files Saved Successfully")
```
