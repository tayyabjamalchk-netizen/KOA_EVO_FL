```python
# ==========================================================
# SECTION 4 : DATASET PATH
# ==========================================================

DATASET_PATH = "/content/Original Images/Original Images/FOLDS/fold1"

TRAIN_PATH = os.path.join(DATASET_PATH, "Train")

VALID_PATH = os.path.join(DATASET_PATH, "Valid")

TEST_PATH = os.path.join(DATASET_PATH, "Test")

print("="*70)
print("DATASET PATHS")
print("="*70)

print(TRAIN_PATH)

print(VALID_PATH)

print(TEST_PATH)
```

```python
# ==========================================================
# SECTION 5 : IMAGE DATA GENERATOR
# Base Paper Implementation
# ==========================================================

train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    zoom_range=0.2,
    shear_range=0.2,
    horizontal_flip=True,
    fill_mode="nearest"
)

valid_datagen = ImageDataGenerator(

    rescale=1./255

)

test_datagen = ImageDataGenerator(

    rescale=1./255

)

print("="*70)
print("IMAGE DATA GENERATORS CREATED")
print("="*70)
```

```python
# ==========================================================
# SECTION 6 : TRAIN GENERATOR
# ==========================================================

train_generator = train_datagen.flow_from_directory(

    TRAIN_PATH,

    target_size=(224,224),

    batch_size=60,

    class_mode="categorical",

    shuffle=True

)

print("="*70)
print("TRAIN GENERATOR CREATED")
print("="*70)
```

```python
# ==========================================================
# SECTION 7 : VALIDATION GENERATOR
# ==========================================================

validation_generator = valid_datagen.flow_from_directory(

    VALID_PATH,

    target_size=(224,224),

    batch_size=60,

    class_mode="categorical",

    shuffle=False

)

print("="*70)
print("VALIDATION GENERATOR CREATED")
print("="*70)
```

```python
# ==========================================================
# SECTION 8 : TEST GENERATOR
# ==========================================================

test_generator = test_datagen.flow_from_directory(

    TEST_PATH,

    target_size=(224,224),

    batch_size=60,

    class_mode="categorical",

    shuffle=False

)

print("="*70)
print("TEST GENERATOR CREATED")
print("="*70)
```

```python
# ==========================================================
# SECTION 9 : CLASS INDICES
# ==========================================================

print("="*70)
print("CLASS INDICES")
print("="*70)

print(train_generator.class_indices)
```

```python
# ==========================================================
# SECTION 10 : PREPROCESSING COMPLETED
# ==========================================================

print("="*70)
print("MONKEYPOX PREPROCESSING COMPLETED SUCCESSFULLY")
print("="*70)

print(f"Train Images      : {train_generator.samples}")
print(f"Validation Images : {validation_generator.samples}")
print(f"Test Images       : {test_generator.samples}")

print("\nDataset Ready For VGG16 Training.")
```
